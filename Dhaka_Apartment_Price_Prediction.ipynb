```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df1 = pd.read_csv('/content/Dhaka_Apartment_Dataset.csv')
df2 = pd.read_csv('/content/bproperty_data.csv')
```

```python
df2.head()
```

```python
len(df2['Region'].unique())
```

```python
df3 = df2[['Price', 'Type', 'No. Beds', 'No. Baths', 'Area', 'Region']]
df3.head(1)
```

```python
df4 = df3.rename(columns={
    'No. Beds': 'Bed',
    'No. Baths': 'Bath',
    'Area' : 'Size',
    'Region': 'Location'
})
df4.head(1)
```

```python
df5 = df4.iloc[:,[0,1,4,2,3,5]]
df5.head(1)
```

```python
df5.dtypes
```

```python
df5.isnull().sum()
```

```python
df6 = df5.dropna()
df6.head(1)
```

```python
df6.isnull().sum()
```

```python
df6.head()
```

```python
df6.Bed = df6.Bed.astype(int)
df6.Bath = df6.Bath.astype(int)
```

```python
df6.dtypes
```

```python
df6.head()
```

```python
df6.shape
```

```python
df1.head(1)
```

```python
df7 = df1[['price' , 'size_in_sqf', 'number_of_bath','number_of_bed','location_area','type_of_listing']]
df7.head(1)
```

```python
df8 = df7.rename(columns={
    'price': 'Price',
    'size_in_sqf': 'Size',
    'number_of_bath': 'Bath',
    'number_of_bed': 'Bed',
    'location_area': 'Location',
    'type_of_listing': 'Type'
})
df8.head(1)
```

```python
df9 =df8.iloc[:,[0,5,1,3,2,4]]
df9.head()
```

```python
df6.dtypes
```

```python
df9.dtypes
```

```python
df9.Price = df9.Price.astype(int)
df9.dtypes
```

```python
df10 = pd.concat([df6,df9],axis = 0)
df10.head()
```

```python
df10.shape
```

```python
df10.to_csv('final_data.csv')
```
